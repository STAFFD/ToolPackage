A block of unknown material is weighed and the mass is determined. The block is then tested in a calorimetry experiment. In this experiment the block is placed in a thermal reservoir. The block's temperature increases, until it reaches the thermal reservoir's temperature. The amount of heat required to bring the block to the reservoir's temperature is recorded. This is repeated with thermal reservoirs of different temperature. The block's change in temperature is also recorded.

Plot Q vs ΔT, including units and labels. Find the block's specific heat from the plot created. Determine the block's material by comparing the experimental specific heat to a table of known specific heats:

```python
a='''
0	0
10	249.45
20	510.47
30	758.43
40	1018.08
50	1240.78
60	1498.48
70	1775.94
80	1992.98
90	2276.90
100	2557.53
'''
dT = []
Q = []
for each in a.split('\n'):
    parts = each.split("\t")
    if len(parts) > 1:
        dT.append(parts[0])
        Q.append(parts[1])
Q = [float(each) for each in Q]
dT = [int(each) for each in dT]
m = 32 # 32g

print(dT)
print(Q)

from plts.func.plot2D import simplePlot, FitPlot
simplePlot(Q, dT, x_label="ΔT(K)", y_label="Q(J)", fitPlot=FitPlot("linear", forceZero=True))
```
